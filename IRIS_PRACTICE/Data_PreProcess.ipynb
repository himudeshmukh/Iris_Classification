```python
df = pd.read_csv('imports-85.data',header=None,na_values="?" )
df.head(5)
```

```python
headers = ["symboling", "normalized_losses", "make", "fuel_type", "aspiration",
           "num_doors", "body_style", "drive_wheels", "engine_location",
           "wheel_base", "length", "width", "height", "curb_weight",
           "engine_type", "num_cylinders", "engine_size", "fuel_system",
           "bore", "stroke", "compression_ratio", "horsepower", "peak_rpm",
           "city_mpg", "highway_mpg", "price"]
```

```python
df.columns=headers
```

```python
df.info()
```

```python
df_temp=df.dropna()   #drop rows     # important data loss    datasize is reduced
df_temp.head(5)
```

```python
df["normalized_losses"].fillna(1,inplace=True)    # replacing by a particular  value
df.info()
```

```python
df["bore"].fillna(df.bore.mean(),inplace=True)   #replacing mean,median,mode
df.info()
```

```python
from scipy.stats import mode
mode(df['num_doors'])
```

```python
mode(df['num_doors']).mode[0]
```

```python
df['num_doors'].fillna(mode(df['num_doors']).mode[0],inplace=True)
df.info()
```
